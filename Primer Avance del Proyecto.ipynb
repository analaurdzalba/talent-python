```python
print("Talento INT Corporativo")
print("Ana Laura Rodríguez Alba")
print("Menú de opciones")
print("1. Alta empleados")
print("2. Crear archivo de emails")
print("3. Reporte de empleados")
print("4. Reporte de cuentas de correo")
print("5. Salir")
opcion = input("Teclea la opción que deseas activar para dar de alta al archivo: ")
while opcion != "5":
    if opcion == "1":
        fprimerArchivo = open('TalentoINT.txt',"a")
        i = 1
        for i in range(1,10000):
            sID = input("Teclea la identificación del empleado: ")
            sNombre = input("Teclea el nombre completo del empleado: ")
            carrera = input("Teclea la carrera del empleado: ")
            direccion = input("Teclea la dirección del empleado: ")
            ciudad = input("Teclea la ciudad del empleado: ")
            telCelular = input("Teclea el celular del empleado: ")
            puesto = input("Teclea el puesto del empleado en la empresa: ")
            persona = (sID +("-")+sNombre +("-")+ carrera +("-")+ direccion +("-")+ ciudad+("-")+telCelular+("-")+puesto+"\n")
            print(persona)
            fprimerArchivo.write(sID +("-")+sNombre +("-")+ carrera +("-")+ direccion +("-")+ ciudad+("-")+telCelular+("-")+puesto+"\n")
            seguir = input("Teclea 1 si deseas continuar registrando empleados, 2 para salir: ")
            if seguir == '2':
                fprimerArchivo.close()
                break
            elif seguir == '1':
                i += 1
        fprimerArchivo.close()
    else:
        if opcion == "2":
            primerArchivo2 = open("TalentoINT.txt", "r")
            segundoArchivo = open("emailTalentoINT.txt", "a")
            for line in primerArchivo2:
                s = line
                posicion = s.find('-')
                posicionDos = s.find('-',posicion+1)
                nombre = s[posicion+1:posicionDos]
                otroNombre = nombre
                nombre = nombre.replace("ñ","n")
                nombre = nombre.replace("á","a")
                nombre = nombre.replace("é","e")
                nombre = nombre.replace("í","i")
                nombre = nombre.replace("ó","o")
                nombre = nombre.replace("ú","u")
                nombre = nombre.replace(".","")
                nombre = nombre.split()
                correo = ""
                for sPedazo in nombre:
                    correo = correo + sPedazo[:3]+"."
                correo = correo.lower()[:-1]+"@int.mx"
                distancia = len(otroNombre)
                espaciado = 30-distancia
                enBlanco = espaciado*(' ')
                segundoArchivo.write(otroNombre+enBlanco+correo+"\n")
                print(otroNombre+enBlanco+correo)
            segundoArchivo.close()
            primerArchivo2.close()
        else: 
            if opcion == "3":
                lineas = 0
                primerArchivo3 = open('TalentoINT.txt',"r")
                print("Reporte de Talento INT")
                print("Monterrey,N.L.")
                import datetime
                today = datetime.date.today()
                print(today)
                print("ID                    Nombre                    Carrera                    Celular                    Puesto")
                for line in primerArchivo3:
                    m = line
                    lugar = m.find('-')
                    iD = m[:posicion]
                    segundoLugar = m.find('-',lugar+1)
                    nom = m[lugar+1:segundoLugar]
                    tercerLugar = m.find('-',segundoLugar+1)
                    car = m[segundoLugar+1:tercerLugar]
                    cuartoLugar = m.find('-',tercerLugar+1)
                    quintoLugar = m.find('-',cuartoLugar+1)
                    sextoLugar = m.find('-',quintoLugar+1)
                    celular = m[quintoLugar+1:sextoLugar]
                    puesto = m[sextoLugar+1:-1]
                    longitudID = len(iD)
                    longitudNom = len(nom)
                    longitudCar = len(car)
                    longitudCel = len(celular)
                    longitudPuesto = len(puesto)
                    distanciaID = 22-longitudID
                    distanciaNom = 26-longitudNom
                    distanciaCar = 27-longitudCar
                    distanciaCel = 27-longitudCel
                    distanciaPuesto = 26-longitudPuesto
                    espacioID = distanciaID*(' ')
                    espacioNom = distanciaNom*(' ')
                    espacioCar = distanciaCar*(' ')
                    espacioCel = distanciaCel*(' ')
                    print(iD+espacioID+nom+espacioNom+car+espacioCar+celular+espacioCel+puesto+"\n")
                    lineas += 1
                    numero = int(lineas)
                    stringYa = str(numero)
                    if lineas % 10 == 0:
                        seguir = input("Teclea 1 para ver más empleados: ")
                        if seguir == '1':
                            lineas == 0
                else:
                    print("La cantidad de empleados de la empresa es de "+ stringYa)
                primerArchivo3.close()
            else:
                if opcion == "4":
                    lineasHola = 0
                    print("Cuentas de correo de Talento INT")
                    print("Monterrey,N.L.")
                    import datetime
                    today = datetime.date.today()
                    print(today)
                    print("Nombre                        email")
                    segundoArchivo4 = open("emailTalentoINT.txt","r")
                    for line in segundoArchivo4:
                        print(line)
                        lineasHola += 1
                        if lineasHola % 10 == 0:
                            seguir = input("Teclea 1 para ver más empleados: ")
                            if seguir == '1':
                                lineasHola == 0
                            else:
                                break
                    segundoArchivo4.close()
    opcion = input("Teclea la opción que deseas activar para dar de alta al archivo: ")
print("Se he cerrado el programa.")
```
